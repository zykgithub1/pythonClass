```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler().fit(Xtrain.iloc[:,[*range(23),-1]])
```

```python
Xtrain_=Xtrain.copy()
Xtest_=Xtest.copy()
```

```python
Xtrain_.iloc[:,[*range(23),-1]]=ss.fit_transform(Xtrain.iloc[:,[*range(23),-1]])
```

```python
Xtest_.iloc[:,[*range(23),-1]]=ss.fit_transform(Xtest.iloc[:,[*range(23),-1]])
```

```python
from sklearn.linear_model import LogisticRegression as logiR
import pandas as pd
```

```python
Ytrain
```

```python
for i in range(3):
    logi=logiR(solver="newton-cg",max_iter=100**20,multi_class="multinomial").fit(Xtrain_,Ytrain.iloc[:,i].ravel())
    print(y.columns[i])
    print("\tTrain:{}".format(logi.score(Xtrain_,Ytrain.iloc[:,i].ravel())))
    print("\tTest:{}".format(logi.score(Xtest_,Ytest.iloc[:,i].ravel())))
    coeff=pd.DataFrame(logi.coef_).T
    if i!=2:
        coeff["mean"]=coeff.iloc[:,:-1].mean(axis=1)
        coeff["name"]=Xtrain.columns
        coeff.columns=["Average","High","Low","mean","name"]
    else:
        coeff["name"]=Xtrain.columns
        coeff.columns=["Coef","mean"]
    print(coeff.sort_values(by="mean",ascending=False).head())
    print("\t")
        
```
