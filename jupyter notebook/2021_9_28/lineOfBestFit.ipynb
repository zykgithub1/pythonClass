```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import pandas as pd
import numpy as np
```

```python
x1=np.linspace(0,10,50)
x1
```

```python
y=2*x1+np.random.randn(50)*10
plt.scatter(x1,y)
```

```python
data=pd.DataFrame({"x":x1,"y":y})
gridobj=sns.lmplot("x","y",data=data)
plt.xlabel("yes")
# plt.xticks(ticks=np.linspace(0,10,11))
```

```python
x2=[0]*10+[1]*40
data=pd.DataFrame({"x1":x1,"y":y,"x2":x2})
gridobj=sns.lmplot("x1","y",data=data,hue="x2")
```
