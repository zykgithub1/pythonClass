```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeClassifier
```

```python
titan = pd.read_csv("./train.csv")
titan
```

# PassengerId	Pclass	Name	Sex	Age	SibSp	Parch	Ticket	Fare	Cabin	Embarked


```python
x=titan[["Pclass","Age","Sex"]]
y=titan["Survived"]
y
```

## fill up lost data

```python
x["Age"].fillna(value=titan["Age"].mean(),inplace=True)
```

```python
x
```

## data spliting

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=22,test_size=0.2)
```

```python
x_train=x_train.to_dict(orient="records")
x_test=x_test.to_dict(orient="records")
x_train
```

```python
x_test
```

```python
transfer=DictVectorizer()
x_train=transfer.fit_transform(x_train)
x_test=transfer.fit_transform(x_test)
```

```python
estimator=DecisionTreeClassifier(max_depth=15)
estimator.fit(x_train,y_train)
```

```python
y_pre=estimator.predict(x_test)
```

```python
y_pre
```

```python
ret=estimator.score(x_test,y_test)
print(ret)
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(estimator,out_file="tree.doc",feature_names=['Age','Sex',"Pclass=1","Pclass=2"])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param={"n_estimators":[100,120,300],"max_depth":[3,7,11]}
gc=GridSearchCV(rf,param_grid=param,cv=3)
gc.fit(x_train,y_train)
```

```python
gc.score(x_test,y_test)
```
